Extracting all info from first six pages

```
# define an empty list, which we will fill with the results:
results = []
```

```
for i in range(6):
  # remember that i starts at zero vs the first page starts at one... so we need to add 1 to i to get to the proper page
  r = requests.get(base_url, params={"page": i+1})
  # if the loop is empty on a page then we stop the loop then and there:
  if len(r.json()) == 0:
    break
  else:
    # the python built-in 'extend' function adds the specified list elements to the end of an existing list:
    results.extend(r.json())
```
